```python
max_iters = 100
gamma = 0.1
initial_w = np.ones((tx.shape[1]))

plt.title('Logistic regression with GD')
w, loss = logistic_regression(y, tx, initial_w, max_iters, gamma, plot_loss = True)
print("Final loss is {:.2f}".format(loss))
```
